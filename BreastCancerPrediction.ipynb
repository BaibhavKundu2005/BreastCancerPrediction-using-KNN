                                             Breast Cancer prediction using KNN

Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing dataset

Dataset is taken from UCI Machine Learning Repository, Kaggle

```python
df = pd.read_csv('/Users/bijit/Downloads/data.csv')
```

```python
df.head()
```

```python
df.info()
```

Dropping unwanted columns

```python
df = df.drop(['Unnamed: 32','id'],axis = 1)
```

```python
df.shape
```

Converting text value in diagnosis column to numerical value

```python
def diagnosis_num(diagnosis):
    if diagnosis == "M":
        return 1
    else:
        return 0
df['diagnosis'] = df['diagnosis'].apply(diagnosis_num)
```

```python
df.head()
```

EDA

```python
sns.lmplot(data = df,x = 'radius_mean',y = 'texture_mean',hue = 'diagnosis')
```

```python
sns.lmplot(data = df,x = 'smoothness_mean',y = 'compactness_mean',hue = 'diagnosis')
```

```python
X = np.array(df.iloc[:, 1:]) 
y = np.array(df['diagnosis']) 
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42) 
```

Using KNN for classificatiom with number of neighbors = 10

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 10) 
knn.fit(X_train, y_train) 
```

```python
knn.score(X_test, y_test)
```

Checking value of n_neighbors for which accuracy is highest

```python
error_rate = []
for i in range(1, 51, 2):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize = (10,6))
plt.plot(error_rate)
plt.xlabel("number of neighbors")
plt.ylabel("error")
plt.show()
```

Since the lowest error is at 5, we take a relatively close value when the error starts to increase

```python
knn = KNeighborsClassifier(n_neighbors = 7) 
knn.fit(X_train,y_train) 
```

Score increased

```python
knn.score(X_test, y_test)
```

```python
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import root_mean_squared_error, r2_score
```

```python
root_mean_squared_error(y_test,pred)
```

Closer the value of r2_score to 1, better the model

```python
r2_score(y_test,pred)
```
